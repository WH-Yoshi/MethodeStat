```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def read_data(file_path, sep=','):
    f = pd.read_csv(file_path, sep=sep)
    f = f.dropna()
    return f


data = read_data('ressources/baseball_players.csv')
poids = data['Weight']
taille = data['Height']
```

```python
plt.scatter(taille, poids)
plt.xlabel('Taille')
plt.ylabel('Poids')
plt.title('Poids en fonction de la taille')
plt.show()
```

```python
def regression_lineaire(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    x_diff = x - x_mean
    y_diff = y - y_mean
    a = np.sum(x_diff * y_diff) / np.sum(x_diff ** 2)
    b = y_mean - a * x_mean
    return a, b

a, b = regression_lineaire(taille, poids)

(a * 178 / 2.54 + b) / 2.20
```

```python
# Poids ind√©pendant
plt.scatter(taille, poids)
plt.plot(taille, a * taille + b, color='red')
plt.xlabel('Taille')
plt.ylabel('Poids')
plt.title('Poids en fonction de la taille')
plt.show()
```

```python
# Taille en fonction du poids
a, b = regression_lineaire(poids, taille)
plt.scatter(poids, taille)
plt.plot(poids, a * poids + b, color='red')
plt.xlabel('Poids')
plt.ylabel('Taille')
plt.title('Taille en fonction du poids')
plt.show()
```
