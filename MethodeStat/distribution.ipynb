```python
from scipy.stats import binom
import matplotlib.pyplot as plt

# Parameters
p = 0.2
N = 10

# Generate data
data = binom.pmf(k=range(N+1), n=N, p=p)
print(data)

# Plot
plt.bar(range(N+1), data)
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.show()
```

```python
from scipy.stats import binom

# Parameters
p = 0.5
n = 4

prob_at_least_1_boy = 1 - binom.pmf(0, n, p)

prob_exactly_2_boys = binom.pmf(2, n, p)

prob_at_most_2_boys = binom.cdf(2, n, p)

prob_1_or_2_girls = binom.pmf(3, n, p) + binom.pmf(2, n, p)

prob_no_girls = binom.pmf(4, n, p)

print(f"Probability of at least 1 boy: {prob_at_least_1_boy:.4f}")
print(f"Probability of exactly 2 boys: {prob_exactly_2_boys:.4f}")
print(f"Probability of at most 2 boys: {prob_at_most_2_boys:.4f}")
print(f"Probability of 1 or 2 girls: {prob_1_or_2_girls:.4f}")
print(f"Probability of no girls: {prob_no_girls:.4f}")
```
