```python
import random

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy import mean

random.seed(2003)
np.random.seed(2003)
```

```python
moyenne = 68.0
ecart_type = 3.0
nombre_etudiants = 3000
```

```python
pop_df = pd.DataFrame({'poids': np.random.normal(moyenne, ecart_type, nombre_etudiants)})
pop_df = pop_df.round({"poids": 1})
pop_df.head()
```

```python
plt.plot(pop_df['poids'], 'o')
plt.xlabel('Étudiant')
plt.ylabel('Poids (kg)')
plt.title('Poids des étudiants')
plt.show()
```

## Échantillonage
non-exhaustif

```python
n_echantillon = 3000

moyenne_liste = []
taille_echantillon = 25

for i in range(n_echantillon):
    echantillon = random.sample(list(pop_df["poids"]), taille_echantillon)
    moyenne_liste.append(mean(echantillon))
    
poids_moyen_df = pd.DataFrame({'moyenne': moyenne_liste})
```

```python
plt.plot(poids_moyen_df['moyenne'], 'o')
plt.show()
```

```python
poids_moyen_df["moyenne"].std(ddof=0)
```

```python
poids_moyen_df["moyenne"].mean()
```

Exhaustif

```python
n_echantillon = 3000

moyenne_liste = []
taille_echantillon = 25

for i in range(n_echantillon):
    echantillon = random.choices(list(pop_df["poids"]), k=taille_echantillon)
    moyenne_liste.append(mean(echantillon))

poids_moyen_df = pd.DataFrame({'moyenne-exhaustive': moyenne_liste})
```
