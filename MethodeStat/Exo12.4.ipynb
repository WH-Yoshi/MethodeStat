```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(4242)
n = 50000
population = np.random.normal(loc=0, scale=1, size=n)

taille_echantillon = 5
n_echantillons = 500
samples = [np.random.choice(population, taille_echantillon, replace=False) for _ in range(1, n_echantillons-1)]
sample = np.sort(samples)

s_squared_list = []
s_hat_squared_list = []

for sample in samples:
    s_squared = np.var(sample, ddof=1)
    s_hat_squared = (len(sample) / (len(sample) - 1)) * s_squared
    s_squared_list.append(s_squared)
    s_hat_squared_list.append(s_hat_squared)

plt.figure(figsize=(12, 6))

plt.hist(s_squared_list, bins=30, color='blue', label='$s^2$')
plt.axvline(np.mean(s_squared_list), color='blue', linestyle='dashed', linewidth=1.5, label='Moyenne')
plt.hist(s_hat_squared_list, bins=30, color='green', label='$\\hat{s}^2$')
plt.axvline(np.mean(s_hat_squared_list), color='green', linestyle='dashed', linewidth=1.5, label='Moyenne')
plt.ylabel("Fr√©quence")
plt.legend()

plt.tight_layout()
plt.show()
```
