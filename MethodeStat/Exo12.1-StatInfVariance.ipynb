```python
# Population gaussienne
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 0, 1
pop = np.random.normal(mu, sigma, 10000)
pop
```

```python
# plot
plt.hist(pop, bins=100, color='blue', alpha=0.7)
```

```python
# Génération de 1000 échantillons de 3 individus
n = 3
samples = np.random.choice(pop, (1000, n), replace=False)
samples
```

```python
# Moyenne des échantillons
moy_echantillons = []
for individus in samples:
    somme = 0
    for i in individus:
        somme += i
    moyenne = somme / n
    moy_echantillons.append(moyenne)

moy_echantillons
```

```python
# plot
plt.hist(moy_echantillons, bins=100, color='green', alpha=0.7)
```

```python
# Histogramme de la moyenne des variances
var_echantillons = []

for individus in samples:
    # Moyenne d'abord
    somme = 0
    for i in individus:
        somme += i
    moyenne = somme / n
    
    # Variance ensuite
    somme = 0
    for i in individus:
        somme += (i - moyenne)**2
    variance = somme / n  # n-1 pour la variance des échantillons, n pour la variance de la population totale
    var_echantillons.append(variance)

moy_var_echantillons = np.mean(var_echantillons)
moy_var_echantillons
```
