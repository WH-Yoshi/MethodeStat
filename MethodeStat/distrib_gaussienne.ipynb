```python
from scipy.stats import norm
from scipy.stats import binom

import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 500
mu = 151
sigma = 15
```

```python
# A
x = 120
y = 155

p = norm.cdf(y, loc=mu, scale=sigma) - norm.cdf(x, loc=mu, scale=sigma)
print(f"Probabilité que la taille soit comprise entre {x} et {y} cm: {p:.4f}"
      f"\nNombre de personnes entre {x} et {y} cm: {p * N:.2f}")
```

```python
# B
x = 185

p = 1 - norm.cdf(x, mu, sigma)
print(f"Probabilité que la taille soit supérieure à {x} cm: {p:.4f}"
      f"\nNombre de personnes supérieures à {x} cm: {p * N:.2f}")
```

```python
# plot
x = np.linspace(0, 200, 1000)
y = norm.pdf(x, mu, sigma)
plt.plot(x, y)
plt.fill_between(x, y, where=(x >= 120) & (x <= 155), alpha=0.5)
plt.fill_between(x, y, where=x >= 185, alpha=0.5)
plt.show()
```

```python
# Exercice 7

p = 0.5
N = 10
x = 3
y = 6
```

```python
# A
p = binom.cdf(y, N, p) - binom.cdf(x - 1, N, p)
print(f"Probabilité d'obtenir un nombre de faces compris entre {x} et {y} (bornes incluses) en utilisant une loi binomiale: {p:.4f}")
```

```python
# plot
mu = N * p
sigma = np.sqrt(N * p * (1 - p))
axe_x = np.linspace(0, 10, 11)
axe_y = norm.pdf(x, mu, sigma)
plt.plot(axe_x, axe_y)
plt.show()
```

```python
# B
mu = N * p
sigma = np.sqrt(N * p * (1 - p))
p = norm.cdf(y + 0.5, mu, sigma) - norm.cdf(x - 0.5, mu, sigma)
print(f"Probabilité d'obtenir un nombre de faces compris entre {x} et {y} (bornes incluses) en utilisant une approximation normale de la loi binomiale: {p:.4f}")
```
