```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

def read_data(file_path, sep=';'):
    return pd.read_csv(file_path, sep=sep)

data = read_data('ressources/CourbePoidsTaille.csv')
poids = data['Poids']
taille = data['Taille']
```

```python
# Repr√©senter le nuage de points correspondant 

plt.scatter(taille, poids)
plt.xlabel('Taille')
plt.ylabel('Poids')
plt.title('Poids en fonction de la taille')
plt.show()
```

```python
# Construire les droites des moindres carr√©s qui ajuste ces donn√©es, en utilisant respectivement le poids et la taille comme variable ind√©pendante  

def linear_regression(x, y):
    x = x.values.reshape(-1, 1)
    y = y.values.reshape(-1, 1)
    model = LinearRegression().fit(x, y)
    return model

print(linear_regression(taille, poids))

def plot_regression(x, y, model):
    plt.scatter(x, y)
    plt.plot(x, model.predict(x.values.reshape(-1, 1)), color='red')
    plt.xlabel('Taille')
    plt.ylabel('Poids')
    plt.title('Poids en fonction de la taille')
    plt.show()
    
model_taille = linear_regression(taille, poids)
plot_regression(taille, poids, model_taille)
```
