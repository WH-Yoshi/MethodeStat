```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


def read_data(file_path, sep=';'):
    f = pd.read_csv(file_path, sep=sep)
    f = f.dropna()
    return f


data = read_data('ressources/Bacterie.csv')
heure = data['Heure']
pop = data['population']
```

```python
plt.scatter(heure, pop)
plt.xlabel('Heure')
plt.ylabel('population')
plt.title('Population en fonction des heures')
plt.grid(True)
plt.show()
```
