```python
def factorielle(n):
  if n<2:
    return 1
  else:
    return n * factorielle(n-1)
```

```python
factorielle(10)
```

```python
factorielle(1000) # même si nombre entier très très grand et très largement
                  # au-delà de ce que permet un entier 16 ou 32 bits...
                  # toujours OK et relativement rapide
```

Interception plus douce des exceptions et gestion minimale:

Voir: https://docs.python.org/3/tutorial/errors.html

```python
for n in range(1000,10000,1000):
    try:
        print(n,factorielle(n))
    except Exception as inst:
        print(type(inst))
        print(inst.args)
        print(inst)
        print('Error on n:{}'.format(n))
```
