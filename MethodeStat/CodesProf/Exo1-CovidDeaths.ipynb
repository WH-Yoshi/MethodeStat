```python
plt.plot(df['DEATHS'],'.')
plt.hlines(y=df['DEATHS'].max(), xmin=0, xmax=df.shape[0],color='black',linestyles='dotted')
plt.hlines(y=df['DEATHS'].mean(), xmin=0, xmax=df.shape[0],color='orange',linestyles='dashed')
plt.hlines(y=df['DEATHS'].median(), xmin=0, xmax=df.shape[0],color='green',linestyles='dashdot')
plt.show()
```

```python
plt.plot(df['DATE'],df['DEATHS'],'.')
#plt.xticks(rotation='vertical')
plt.xticks(df.iloc[range(0,df.shape[0],100)]['DATE'], rotation=45, ha='right')
plt.hlines(y=df['DEATHS'].max(), xmin=0, xmax=df.shape[0],color='black',linestyles='dotted')
plt.hlines(y=df['DEATHS'].mean(), xmin=0, xmax=df.shape[0],color='orange',linestyles='dashed')
plt.hlines(y=df['DEATHS'].median(), xmin=0, xmax=df.shape[0],color='green',linestyles='dashdot')
plt.show()
```
