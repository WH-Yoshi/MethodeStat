```python
import pandas as pd
import matplotlib.pyplot as plt
```

Import des données

```python
df = pd.read_csv('./chiffresVente.csv', sep=';', parse_dates=True)
```

A quoi ressemble les données?

```python
df.head()
```

Graphe simple des données 'ChiffreVente'. Indice horizontal = position dans le tableau

```python
plt.plot(df['ChiffreVente'], 'o')
plt.show()
```

Créons des étiquettes de date = Mois + Année (concaténation)

```python
df['date'] = df.apply(lambda x:'{} {}'.format(x['Mois'], x['Année']), axis=1)
df
```

Graphe simple des données 'ChiffreVente'. Indice horizontal = étiquette de date => pas beau

```python
plt.plot(df['date'], df['ChiffreVente'], 'o')
plt.show()
```

Graphe simple des données 'ChiffreVente'. Indice horizontal = étiquette de date
+ rotation à 45 degré des étiquettes + justification à droite => trop d'étiquettes!!!

```python
plt.plot(df['date'], df['ChiffreVente'], 'o')
plt.xticks(df['date'], rotation=45, ha='right')
plt.show()
```

```python
plt.plot(df['date'], df['ChiffreVente'], 'o')
plt.xticks(df.iloc[range(0,df.shape[0],3)]['date'], rotation=45, ha='right')
plt.ylabel('Chiffres Vente')
plt.show()
```

Calcul de la moyenne

```python
moyenne_chiffre_vente = df['ChiffreVente'].mean()
moyenne_chiffre_vente
```

Insertion d'une ligne de moyenne dans le graphe précédent

```python
plt.plot(df['date'], df['ChiffreVente'], 'o')
plt.xticks(df.iloc[range(0,df.shape[0],3)]['date'], rotation=45, ha='right')
plt.ylabel('Chiffres Vente')
plt.axhline(y=moyenne_chiffre_vente, color='orange', linestyle='dotted')
plt.show()
```

Calcul + représentation de l'histogramme

```python
df['ChiffreVente'].hist()
plt.xlabel('Chiffre Vente par Mois')
plt.ylabel('#Mois')
plt.show()
```

Ajout de la ligne de moyenne sur le graphe d'histogramme

```python
df['ChiffreVente'].hist()
plt.xlabel('Chiffre Vente par Mois')
plt.ylabel('#Mois')
plt.axvline(x=moyenne_chiffre_vente, color='orange', linestyle='dotted', linewidth=5)
plt.show()
```
