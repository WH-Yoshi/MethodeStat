```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def read_data(file_path, sep=';'):
    return pd.read_csv(file_path, sep=sep)
```

```python
df = read_data('./ressources/corona_cases_belgium.csv')

plt.figure(figsize=(20, 10))
plt.plot(df['CASES'],'.', label='Nombre de cas')
plt.axhline(y=np.mean(df['CASES']), color='red', linestyle='-', label='Moyenne')
plt.hlines(np.median(df['CASES']), 0, df.shape[0], color='green', label='MÃ©diane') 
plt.hlines(np.quantile(df['CASES'], 0.25), 0, df.shape[0], color='purple', label='Q1')
plt.hlines(np.quantile(df['CASES'], 0.75), 0, df.shape[0], color='blue', label='Q3')
plt.legend()
plt.title('Nombre de cas COVID-19 en Belgique')
plt.show()

# Histogram
plt.hist(df['CASES'], bins=15, edgecolor='black')
plt.vlines(df['CASES'].mean(), 0, 900, color='red', label='Moyenne')
plt.title('Histogramme cas COVID-19 en Belgique')
plt.xlabel('Nombre de cas')
plt.ylabel('Jours')
```

```python
data = read_data('./ressources/ChiffresVente.csv')

moyenne_vente = data['ChiffreVente'].mean()

plt.figure(figsize=(12, 6))
plt.plot(data['Mois'] + ' ' + data['AnnÃ©e'].astype(str), data['ChiffreVente'], marker='.')
plt.title('Chiffres de vente par mois (2009-2011)')
plt.xlabel('Mois et AnnÃ©e')
plt.ylabel('Chiffre de Vente')
plt.xticks(rotation=90)
plt.axhline(moyenne_vente, color='r', linestyle='--', label=f'Moyenne: {moyenne_vente:.2f}')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.hist(data['ChiffreVente'], bins=15, color='blue', alpha=0.7)
plt.title('Histogramme des chiffres de vente (2009-2011)')
plt.xlabel('Chiffre de Vente')
plt.ylabel('FrÃ©quence')
plt.axvline(moyenne_vente, color='r', linestyle='--', label=f'Moyenne: {moyenne_vente:.2f}')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
