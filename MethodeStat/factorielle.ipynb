```python
def factorielle(n):
    if n<2:
        return 1
    else:
        return n * factorielle(n-1)
    
print(factorielle(10))
```

```python
import sys
sys.set_int_max_str_digits(10000)

for n in range(2900, 3010, 10):
    try:
        print(n, factorielle(n))
    except RecursionError as e:
        print(type(e))
        print(n, e)
        print(f"error on n: {n}")
```
