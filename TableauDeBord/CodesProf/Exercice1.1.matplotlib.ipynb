```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ocean_dict = {'Pacifique':183.4,'Atlantique':106.7,
              'Indien':73.8,
              'Antarctique':19.7,'Arctique':12.4}
```

```python
ocean_df = pd.DataFrame()

ocean_df['Ocean'] = ocean_dict.keys()
ocean_df['Superficie'] = ocean_dict.values()
```

```python
ocean_df
```

```python
plt.barh(ocean_df['Ocean'], ocean_df['Superficie'], align='center')
plt.gca().invert_yaxis()
plt.title('Superficie des océans')
plt.xlabel('Million km2')
plt.show()
```

```python
plt.figure(figsize=[6,6]) 

plt.title(r'Superficie des océans en millions de km$^{2}$')

labels = ocean_df.apply(lambda x:'{}:{}'.format(x['Ocean'],x['Superficie']),axis=1)
plt.pie(ocean_df['Superficie'], labels=labels, startangle=90)

plt.show()
```
