```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
input_df = pd.read_csv('resources/chiffresvente.csv',sep=';')
```

```python
input_df.head()
```

```python
width=0.5
plt.bar(input_df.index, input_df['2009'], width, label='2009')
plt.show()
```

```python
width=1
i_annee=0
for annee in ['2009', '2010', '2011']:
    #plt.bar(input_df.index +i_annee*width/3, input_df[annee], width/3, label=annee)
    plt.bar(input_df.index+i_annee*width/3 , input_df[annee], width/3, label=annee)
    i_annee += 1
plt.ylabel('Chiffre de vente (MEuros)')
plt.xlabel('Index du mois')
plt.legend()
plt.show()
```

```python
width=1
i_annee=1
for annee in ['2009', '2010', '2011']:
    plt.bar(input_df.index +i_annee*width/3, input_df[annee], width/3, label=annee)
    i_annee += 1
plt.ylabel('Chiffre de vente (MEuros)')
plt.xticks(range(0,12), input_df['Mois'].unique(), rotation=45)
plt.legend()
plt.show()
```

```python
for annee in ['2009','2010','2011']:
    plt.plot(input_df['Mois'],input_df[annee],label=annee)
plt.legend()
plt.xticks(range(0,12), input_df['Mois'].unique(), rotation=45)
plt.ylabel('Chiffre de Vente (MEuros)')
plt.show()
```

```python
transformed_input_df = pd.DataFrame()

sub_df_map = {}
for year in ['2009','2010','2011']:
    sub_df = input_df[['Mois',year]].copy()
    sub_df['Annee'] = year
    sub_df.rename(columns={year:'Vente'}, inplace=True)
    sub_df_map[year] = sub_df
    
transformed_df = pd.concat(sub_df_map.values(), ignore_index=True)
transformed_df
```

```python
plt.plot(transformed_df['Vente'])
plt.show()
```

```python
for annee in ['2009','2010','2011']:
    sub_df = transformed_df[transformed_df['Annee']==annee]
    plt.plot(sub_df['Vente'],label=annee)
plt.legend()
plt.show()
```
