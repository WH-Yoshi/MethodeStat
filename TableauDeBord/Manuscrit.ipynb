```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Charger les données
df = pd.read_csv('ressources/neuralNetworksOnMNIST.csv')

# Ajouter une colonne pour indiquer si la prédiction est correcte
df['correct'] = df['digit'] == df['predicted_digit']

# Filtrer les erreurs
errors = df[df['correct'] == False]

# Nombre total d'erreurs
total_errors = len(errors)
```

```python
# Matrice de confusion
from sklearn.metrics import confusion_matrix
import numpy as np

cm = confusion_matrix(df['digit'], df['predicted_digit'])

# Mettre des zéros sur la diagonale
np.fill_diagonal(cm, 0)

plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.arange(10), yticklabels=np.arange(10))
plt.title('Matrice de confusion avec zéros sur la diagonale')
plt.xlabel('Prédiction')
plt.ylabel('Réel')
plt.show()
```

```python
# Taux d'erreur global
error_rate = total_errors / len(df)
print(f"Taux d'erreur : {error_rate * 100:.2f}%")
```
