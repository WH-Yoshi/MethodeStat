```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
df = pd.read_csv('ressources/GDP_vs_Covid19_VaccineDistribution.1.csv')

# Normalize population for bubble sizes
population_size = df['population'] / 1e6  # Scale down population for better visualization

# Create the bubble chart
plt.figure(figsize=(10, 8))

# Bubble plot
scatter = plt.scatter(df['gdp_per_capita'], df['people_vaccinated_per_hundred'],
                      s=population_size,  # Bubble size is proportional to population
                      c=df['population'],  # Color represents population
                      cmap='viridis',      # Color map for population
                      alpha=0.6, edgecolors="w", linewidth=2)

# Add a color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Population Size')

# Labels and title
plt.xlabel('GDP (in billions)')
plt.ylabel('Vaccination Coverage (%)')
plt.title('COVID-19 Vaccination Coverage vs GDP (Bubble Size: Population)')

# Show plot
plt.tight_layout()
plt.show()
```
